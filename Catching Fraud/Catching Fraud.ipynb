```python
import pandas as pd
```

```python
transactions = pd.read_csv(r'C:\Users\Ashish\Downloads\hc_data\transactions.csv')
fraudsters = pd.read_csv(r'C:\Users\Ashish\Downloads\hc_data\fraudsters.csv')
```

```python
fraud_transact.to_csv(r'C:\Users\Ashish\Desktop\fraud_transactions.csv')
```
